Сопоставляем оквэды закупкам

```python
id_parts = parts.merge(comp, on=['supplier_inn'], how='left')
id_parts['okved'].fillna('00', inplace=True)
id_parts = id_parts.groupby('id').agg({'okved': list}).reset_index()
```

```python
parts_comp = parts.merge(comp, on=['supplier_inn'], how='left')
```

выигрышные оквэды

```python
win_parts = parts_comp[parts_comp.is_winner == 'Да']
```

```python
win_id2okved = {}
for i, o in zip(win_parts.id, win_parts.okved):
    win_id2okved[i] = o 
```

```python
win_id2okved
```

```python
from collections import Counter
id_parts['okved'] = id_parts.okved.map(lambda x: Counter([s.split('.')[0] if '.' in s and '?' not in s else '00' for e in x for s in e.split('||')]).most_common())
```

```python
for e in id_parts.okved:
    for x in e:
        if '.' in x[0]:
            print(x)
```

Сопоставляем оквэды по популярности

```python
id2okved = {}
for i, o in zip(id_parts.id, id_parts.okved):
    if len(o) > 1 and o[0][1] != o[1][1]:
        id2okved[i] = o[0][0]
    elif o[0][0] == '00' and len(o) == 1:
        id2okved[i] = o[0][0]
    elif len(o) > 1:
        for x in o:
            if x[0] != '00':
                id2okved[i] = x[0]
                break
    else:
        id2okved[i] = o[0][0] 
```

```python
df
```

```python
Counter(id2okved.values())
```

```python
filter_df = df[df.id.isin(id2okved)]
filter_df['okved'] = filter_df.id.map(id2okved)
```

```python
filter_df['publish_date'] = pd.to_datetime(filter_df.publish_date)
filter_df['month'] = filter_df.publish_date.dt.month
filter_df['year'] = filter_df.publish_date.dt.year
```

Код для обучения катбуста на временных рядах по разным таргетам

```python
train = filter_df.groupby(['okved', 'delivery_region', 'month', 'year']).agg({'price': 'sum'}).reset_index()
```

```python
train.to_csv('train_tender.csv', index=False)
```

```python
from catboost import CatBoostRegressor
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(train.drop(columns=['price']), train['price'], test_size=0.2, random_state=42)

reg = CatBoostRegressor(iterations=1000, loss_function='RMSE', verbose=10, learning_rate=0.05)
reg.fit(X_train, y_train, eval_set=(X_test, y_test), cat_features=['okved', 'delivery_region'])
```

Получение вектора из текста

```python
!pip install transformers sentencepiece
import torch
from transformers import AutoTokenizer, AutoModel
tokenizer = AutoTokenizer.from_pretrained("cointegrated/rubert-tiny2")
model = AutoModel.from_pretrained("cointegrated/rubert-tiny2").cuda()
# model.cuda()  # uncomment it if you have a GPU

def embed_bert_cls(text, model, tokenizer):
    t = tokenizer(text, padding=True, truncation=True, return_tensors='pt')
    with torch.no_grad():
        model_output = model(**{k: v.to(model.device) for k, v in t.items()})
    embeddings = model_output.last_hidden_state[:, 0, :]
    embeddings = torch.nn.functional.normalize(embeddings)
    return embeddings[0].cpu().numpy()


# (312,)
```

```python
train_okved.lot_name.dropna()
```

```python
lot_emb = []
from tqdm.auto import tqdm
for lot_name in tqdm(train_okved.lot_name.dropna()):
    lot_emb.append(embed_bert_cls(lot_name, model, tokenizer))
```

```python
train_emb = train_okved.dropna(subset=['lot_name']).reset_index(drop=True).join(pd.DataFrame(lot_emb, columns=[f"rubert_{i}" for i in range(312)]))
```
